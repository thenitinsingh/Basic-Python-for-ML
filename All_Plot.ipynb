```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# Rename 'class' column to 'passenger_class'
titanic.rename(columns={'class': 'passenger_class'}, inplace=True)

# 1. Line Plot
plt.figure(figsize=(8, 5))
sns.lineplot(x=titanic.index, y='survived', data=titanic)
plt.title('Line Plot')
plt.show()

# 2. Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='passenger_class', y='survived', data=titanic)
plt.title('Bar Plot')
plt.show()

# 3. Histogram
plt.figure(figsize=(8, 5))
sns.histplot(titanic['age'].dropna(), bins=20)
plt.title('Histogram')
plt.show()

# 4. Box Plot
plt.figure(figsize=(8, 5))
sns.boxplot(x='passenger_class', y='age', data=titanic)
plt.title('Box Plot')
plt.show()

# 5. Scatter Plot
plt.figure(figsize=(8, 5))
sns.scatterplot(x='age', y='fare', hue='survived', data=titanic)
plt.title('Scatter Plot')
plt.show()

# 6. Heatmap
plt.figure(figsize=(8, 5))
titanic_numeric = titanic.select_dtypes(include=['float64', 'int64'])
corr_matrix = titanic_numeric.corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Heatmap')
plt.show()

# 7. Violin Plot
plt.figure(figsize=(8, 5))
sns.violinplot(x='sex', y='age', data=titanic, hue='survived', split=True)
plt.title('Violin Plot')
plt.show()

# 8. Pair Plot
plt.figure(figsize=(8, 5))
sns.pairplot(titanic[['age', 'fare', 'pclass', 'survived']].dropna(), hue='survived')
plt.title('Pair Plot')
plt.show()

# 9. Joint Plot
sns.jointplot(x='age', y='fare', data=titanic, kind='scatter')
plt.suptitle('Joint Plot', y=1.03)
plt.show()

# 10. KDE Plot
plt.figure(figsize=(8, 5))
sns.kdeplot(titanic['age'].dropna(), shade=True)
plt.title('KDE Plot')
plt.show()

# 11. Count Plot
plt.figure(figsize=(8, 5))
sns.countplot(x='passenger_class', data=titanic)
plt.title('Count Plot')
plt.show()

# 12. Swarm Plot
plt.figure(figsize=(8, 5))
sns.swarmplot(x='passenger_class', y='age', data=titanic)
plt.title('Swarm Plot')
plt.show()

# 13. Facet Grid
g = sns.FacetGrid(titanic, col='sex', row='survived')
g.map(sns.histplot, 'age')
g.fig.suptitle('Facet Grid', y=1.03)
plt.show()

# 14. Regression Plot
plt.figure(figsize=(8, 5))
sns.regplot(x='age', y='fare', data=titanic.dropna(subset=['age', 'fare']))
plt.title('Regression Plot')
plt.show()

# 15. Stacked Bar Plot
plt.figure(figsize=(8, 5))
titanic_class_survived = pd.crosstab(titanic['passenger_class'], titanic['survived'])
titanic_class_survived.plot(kind='bar', stacked=True, color=['red', 'green'])
plt.title('Stacked Bar Plot')
plt.show()

# 16. Area Plot
plt.figure(figsize=(8, 5))
titanic_sorted_age = titanic['age'].sort_values().dropna()
plt.fill_between(titanic_sorted_age.index, titanic_sorted_age.cumsum(), color="skyblue", alpha=0.4)
plt.title('Area Plot')
plt.show()

# 17. Error Bar Plot
plt.figure(figsize=(8, 5))
sns.barplot(x='passenger_class', y='fare', data=titanic, ci='sd')  # `ci='sd'` adds error bars based on standard deviation
plt.title('Error Bar Plot')
plt.show()

# 18. Bubble Plot
plt.figure(figsize=(8, 5))
plt.scatter(titanic['age'], titanic['fare'], s=titanic['fare']*2, alpha=0.5)
plt.title('Bubble Plot')
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()

# 19. Categorical Plot
plt.figure(figsize=(8, 5))
sns.catplot(x='passenger_class', y='age', hue='sex', kind='swarm', data=titanic)
plt.title('Categorical Plot')
plt.show()
```
